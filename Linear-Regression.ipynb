```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x=np.arange(0,10)
#function to assign a random sequence of numbers
y=[3,4,5,7,9,8,9,10,12,18]
print(x)
plt.scatter(x,y)
plt.show()
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
#sets degree of polynomial function
poly_features=poly.fit_transform(x.reshape(-1,1))
#.reshape(-1,1) transforms x as a 1D array into a 2D array
print(poly_features)
model=LinearRegression()
#polynomial regression is a linear regression model
model.fit(poly_features,y)
y_pred=model.predict(poly_features)
#plt.figure(figsize=(10,6))
plt.title('Polynomial regression model')
plt.scatter(x,y)
plt.plot(x,y_pred,c='red')
plt.show()
```
