```python
# importing libraries
import pandas as pd
import numpy as py
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# reading csv file to df
df = pd.read_csv('datasets/IRIS_DATASET.csv') 

# assingning data to x
x = df.drop('species', axis = 1)

# assigning label to y
y = df['species']

# standardizing the dataset
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# perform PCA
pca_model = PCA(n_components = 2)
x_pca = pca_model.fit_transform(x_scaled)

# Check the shape of the model 
print('original shape = ', x.shape)
print('transformed shape = ', x_pca.shape)

# plotting the model
plt.figure(figsize = (10,10))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = 'r')
plt.xlabel('Principle component 1')
plt.ylabel('Principle Component 2')
plt.title('Principle Component Analysis of Iris dataset')
plt.legend()

# printing the explained variance of the model
explained_variance = pca_model.explained_variance_ratio_
print(f'Explained variance of each component = {explained_variance}')
```
