```python
for i in range(1,10):
    print(f"K = {i}")
    kmeans = KMeans(n_clusters=i, random_state=42, n_init="auto")
    y_kmeans = kmeans.fit_predict(X)
    
    # Evaluate clustering performance using built-in metrics
    rand_index = adjusted_rand_score(y_true, y_kmeans)
    homogeneity = homogeneity_score(y_true, y_kmeans)
    
    print(f'Adjusted Rand Index: {rand_index:.2f}')
    print(f'Homogeneity Score: {homogeneity:.2f}')
    print()

    # Scatter plot to show the clusters
    plt.scatter(df[iris.feature_names[0]], df[iris.feature_names[1]], c=y_kmeans, cmap='viridis', s=50, alpha=0.7)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
    plt.xlabel(iris.feature_names[0])  # x-axis label (e.g., sepal length)
    plt.ylabel(iris.feature_names[1])  # y-axis label (e.g., sepal width)
    plt.title('KMeans Clustering of Iris Dataset')
    plt.legend()
    plt.show()
```
