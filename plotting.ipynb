```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df=pd.read_csv('datasets/STUDENTSCORE_DATASET.csv')
df.head()
x=df.iloc[:,:-1]
print(x)
print()
y=df.iloc[:,1]
print(y)
plt.scatter(x,y,color='red')
plt.title('Linear Regression')
plt.xlabel('X-Axis Hours')
plt.ylabel('Y-Axis Scores')
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.33,random_state=40)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_test,y_pred)
plt.plot(x_test,y_pred,color='blue',linewidth=2,label='Predicted Line')
plt.scatter(x,y,color='red')
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print(f'Mean squared Error:{mse}')
y_pred=model.predict([[5.2]])
print(y_pred)
```
